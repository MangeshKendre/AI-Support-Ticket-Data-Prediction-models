```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters, numbers, and extra whitespaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize the text
    words = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Join the words back into a cleaned text
    cleaned_text = ' '.join(words)

    return cleaned_text
```

```python
df['Description'].apply(preprocess_text).head(33)
```

```python
df['CleanedDescription'] = df['Description'].apply(preprocess_text)
```

```python
# Date Processing
df['Submit Date'] = pd.to_datetime(df['Submit Date'])
df['Last Resolved Date'] = pd.to_datetime(df['Last Resolved Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
df['Responded Date'] = pd.to_datetime(df['Responded Date'])
```

```python
# Ticket Classification
X_text = df['CleanedDescription']
y_category = df['Categorization Tier 1']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_text, y_category, test_size=0.2, random_state=42)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Model for Ticket Classification (Naive Bayes)
classifier = MultinomialNB()
classifier.fit(X_train_tfidf, y_train)
y_pred_category = classifier.predict(X_test_tfidf)
```

```python
y_pred_category,y_category
```

```python
# Evaluate Ticket Classification Model
accuracy_category = accuracy_score(y_test, y_pred_category)
precision_category = precision_score(y_test, y_pred_category, average='weighted')
recall_category = recall_score(y_test, y_pred_category, average='weighted')
f1_category = f1_score(y_test, y_pred_category, average='weighted')

print(f'Ticket Classification Accuracy: {accuracy_category}')
print(f'Ticket Classification Precision: {precision_category}')
print(f'Ticket Classification Recall: {recall_category}')
print(f'Ticket Classification F1 Score: {f1_category}')

# Classification Report
print('\nClassification Report:')
print(classification_report(y_test, y_pred_category))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

def train_and_evaluate_models(X_train, y_train, X_test, y_test):
    models = {
        'Naive Bayes': MultinomialNB(),
        'Random Forest': RandomForestClassifier(),
        'SVM': SVC(),
        'K-Nearest Neighbors': KNeighborsClassifier(),
        'Logistic Regression': LogisticRegression(),
        'Decision Tree': DecisionTreeClassifier()
    }

    results = {}

    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        f1 = f1_score(y_test, y_pred, average='weighted')

        results[model_name] = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        }

        print(f'{model_name} Results:')
        print(f'Accuracy: {accuracy}')
        print(f'Precision: {precision}')
        print(f'Recall: {recall}')
        print(f'F1 Score: {f1}\n')

    return results

# Train and evaluate multiple models
results = train_and_evaluate_models(X_train_tfidf, y_train, X_test_tfidf, y_test)

# Compare model performance
for model_name, metrics in results.items():
    print(f'{model_name} - Accuracy: {metrics["Accuracy"]}, F1 Score: {metrics["F1 Score"]}')
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder

# Create a copy of the DataFrame to avoid modifying the original data
df_encoded = df1.copy()

ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

# Fit and transform the labels in your training data
categorical_columns = ['Categorization Tier 1', 'Categorization Tier 2', 'Categorization Tier 3']
df_encoded[categorical_columns] = ordinal_encoder.fit_transform(df_encoded[categorical_columns])


# Display the encoded DataFrame
df_encoded.head()
```

```python
# Response Time Prediction
X_regression_encoded = df_encoded[[ 'Categorization Tier 1', 'Categorization Tier 2', 'Categorization Tier 3']]
y_regression_encoded = df_encoded['MTTR in Mins']

# Split the data into training and testing sets
X_train_reg_encoded, X_test_reg_encoded, y_train_reg_encoded, y_test_reg_encoded = train_test_split(
    X_regression_encoded, y_regression_encoded, test_size=0.2, random_state=42
)
```

```python
# Model for Response Time Prediction (Random Forest Regressor)
random_forest_regressor_model = RandomForestRegressor()
random_forest_regressor_model.fit(X_train_reg_encoded, y_train_reg_encoded)

# Predict using the trained Random Forest Regressor
y_pred_reg_encoded = random_forest_regressor_model.predict(X_test_reg_encoded)

# Evaluate Response Time Prediction Model
mae_reg_encoded = mean_absolute_error(y_test_reg_encoded, y_pred_reg_encoded)
mse_reg_encoded = mean_squared_error(y_test_reg_encoded, y_pred_reg_encoded)
r2_reg_encoded = r2_score(y_test_reg_encoded, y_pred_reg_encoded)

print('Random Forest Regressor Results:')
print(f'MAE: {mae_reg_encoded}')
print(f'MSE: {mse_reg_encoded}')
print(f'R^2: {r2_reg_encoded}')
```

# 3. Sentiment Analysis

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
# Download necessary resources for NLTK
nltk.download('vader_lexicon')
```

```python
# Sample DataFrame with ticket descriptions and responses
data = {
    'Description': ['The product is excellent!', 'Service was terrible.', 'Issue resolved quickly.'],
    'Response': ['Thank you for your feedback.', 'We apologize for the inconvenience.', 'Glad we could assist.']
}

df = pd.DataFrame(data)

# Initialize SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Function to get sentiment scores
def get_sentiment(text):
    sentiment_score = sia.polarity_scores(text)
    return sentiment_score

# Apply sentiment analysis to ticket descriptions and responses
df['Description_Sentiment'] = df['Description'].apply(get_sentiment)
df['Response_Sentiment'] = df['Response'].apply(get_sentiment)

# Display the DataFrame with sentiment scores
print(df[['Description', 'Description_Sentiment', 'Response', 'Response_Sentiment']])
```

```python
# Take user input for ticket description
user_input_description = input("Enter the ticket description: ")

# Apply sentiment analysis to the user input
user_sentiment = get_sentiment(user_input_description)

# Display the sentiment scores for the user input
print(f'Sentiment Analysis for User Input:')
print(f'Text: {user_input_description}')
print(f'Sentiment Scores: {user_sentiment}')
```
